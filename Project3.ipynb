# Analyzing Used Car Listings on eBay Kleinanzeigen #
We will be working on the dataset of used cars from eBay Kleninzeigen, a [classifeds](https://en.wikipedia.org/wiki/Classified_advertising) section of the Germany eBay website.

The datset was originally scraped and uploaded to [Kaggle](https://www.kaggle.com/orgesleka). the version of the datset we are working with a sample of 50,000 data points that was prepared by Dataquest including simulating a less-cleaned version of the data.

The data dictionary provided with the data is as follows:

- `dateCrawled` - When this ad was first crawled. All field-values are taken from this data.
- `name` - Name of the care.
- `seller` - Whether the seller is private or dealer.
- `offerType` - The type of listing.
- `price` - The price on the ad to sell the car.
- `abtest` - Whether the listing is included in A/B test.
- `vehicleType` - The vehicle type.
- `yearOfRegistration` - The year in which year the car was first registered.
- `gearbox` - The transmission type.
- `powerPS` - The power of the car in PS.
- `model` - The car model name.
- `kilometer` - How many kilometers the car has driven.
- `monthOfRegistraion` - The month in which the car was registered.
- `fuelType` - What type of fuel car uses.
- `brand` - The brand of the car.
- `notRepairedDamage` - If the car has a damage which is not yet repaired.
- `dateCreated` - The date on which the eBay listing was creating.
- `nrOfPictures` - The number of pictures in the ad.
- `postalCode` - The postal code for the location of the vehicle.
- `lastSeenOnline` - When the crawler saw this ad last online.

## Initial Data Exploration and Cleaning ##
We'll start by exploring the data to find obvious areas where we can clean the data.

```python
autos.describe(include = 'all')
```

Our initial observations:
- There are number of text columns where all (or nearly all) of the values are the same:
    - **seller**
    - **offer_type**
- the num_photos columns looks odd, we'll need to investigate this further.    

```python
autos["num_photos"].value_counts()
```

It looks like the **num_photos** column has 0 for every column. We'll drop this column, plus the other two we notes as mostly value one.

```python
autos = autos.drop(["num_photos", "seller", "offer_type"], axis = 1)
```

There are two columns, **price** and **auto**, which are numeric values with extra characters being stored as text. We'll clean and convert these.

```python
autos["price"] = (autos["price"]
                  .str.replace("$", "")
                  .str.replace(",", "")
                  .astype(int)
                 )

autos["price"].head()
```

```python
autos["odometer"] = (autos["odometer"]
                     .str.replace("km", "")
                     .str.replace(",", "")
                     .astype(int)
                    )

autos.rename({"odometer": "odometer_km"}, axis = 1, inplace = True)
autos["odometer_km"].head()
```

## Dealing with Incorrect Registraion Year Data ##

Beacuse a car can't be registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determing the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s.

One option is to remove the listings with these values. Let's determine what percentage of our data has invalid values in this columns:

```python
(~autos["registration_year"].between(1900,2016)).sum() / autos.shape[0]
```

Given that this is less than 4% of our data, We will remove those rows. 

```python
# Many ways to select rows in a dataframe that fall within a value range for a column.
# Using `Series.between()` in one way.
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize = True).head(10)
```

It appears that most of the vehicles were first registered in the past 20 years.

## Exploring Price by Brand ##

```python
autos["brand"].value_counts(normalize = True)
```

German manufacturers represent four out of the top six brands, almost 50% of the overall listings. Volkswagen is by far the most popular brand, with approximately double the cars for sale of the next two brands combined.

There are lots of brands that don't have a significant percentage of listings, so we will limit our analysis to brands representing more than 5% of total listings.

```python
brand_counts = autos["brand"].value_counts(normalize = True)
common_brands = brand_counts[brand_counts > 0.05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)
    
brand_mean_prices    
```

Of the top 6 brands , there is a distinct price gap:
- Audi, BMW, Mercedes Benz are more expensive.
- Ford and Opel are less expensive.
- Volkswagen is in between - this may explain its popularity, it may be a 'best of 'both worlds' option.

```python
bmp_series = pd.Series(brand_mean_prices)
pd.DataFrame(bmp_series, columns = ["mean_price"])
```

## Exploring Mileage ##

```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending = False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending = False)
```

```python
brand_info = pd.DataFrame(mean_mileage, columns = ['mean_mileage'])
brand_info
```

```python
brand_info["mean_price"] = mean_prices
brand_info
```

The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% of the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
